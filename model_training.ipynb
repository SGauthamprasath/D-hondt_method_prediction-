```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV
import joblib
from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error
```

```python
df=pd.read_csv('dataset\GoesGold\ElectionData.csv')
df
```

```python
df.columns
```

```python
training_df=df[['Votes','validVotesPercentage','totalMandates','Hondt']]
training_df
```

```python
target_df=df['FinalMandates']
target_df
```

```python
training_df.isnull().sum(),target_df.isnull().sum()
```

```python
X_train,X_test,y_train,y_test=train_test_split(training_df,target_df,test_size=0.2,random_state=0)
```

```python
scaler=StandardScaler()

X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
joblib.dump(scaler,'scaler.pkl')
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
```

```python
LR_model=LinearRegression()
R_model=Ridge()
L_model=Lasso()
RF_model=RandomForestRegressor()
```

```python
LR_model.fit(X_train_scaled,y_train)
R_model.fit(X_train_scaled,y_train)
L_model.fit(X_train_scaled,y_train)
RF_model.fit(X_train_scaled,y_train)
```

```python
LR_y_pred=LR_model.predict(X_test_scaled)
LR_r2 = r2_score(y_test, LR_y_pred)
LR_mae = mean_absolute_error(y_test, LR_y_pred)
LR_rmse = root_mean_squared_error(y_test, LR_y_pred)
```

```python
print(f"R2 Score: {LR_r2:.4f}")
print(f"MAE: {LR_mae:.4f}")
print(f"RMSE: {LR_rmse:.4f}")
```

```python
R_y_pred=R_model.predict(X_test_scaled)
R_r2 = r2_score(y_test, R_y_pred)
R_mae = mean_absolute_error(y_test, R_y_pred)
R_rmse = root_mean_squared_error(y_test, R_y_pred)
print(f"R2 Score: {R_r2:.4f}")
print(f"MAE: {R_mae:.4f}")
print(f"RMSE: {R_rmse:.4f}")
```

```python
L_y_pred=L_model.predict(X_test_scaled)
L_r2 = r2_score(y_test, L_y_pred)
L_mae = mean_absolute_error(y_test, L_y_pred)
L_rmse = root_mean_squared_error(y_test, L_y_pred)
print(f"R2 Score: {L_r2:.4f}")
print(f"MAE: {L_mae:.4f}")
print(f"RMSE: {L_rmse:.4f}")
```

```python
RF_y_pred=RF_model.predict(X_test_scaled)
RF_r2 = r2_score(y_test, RF_y_pred)
RF_mae = mean_absolute_error(y_test, RF_y_pred)
RF_rmse = root_mean_squared_error(y_test, RF_y_pred)
print(f"R2 Score: {RF_r2:.4f}")
print(f"MAE: {RF_mae:.4f}")
print(f"RMSE: {RF_rmse:.4f}")
```

```python
joblib.dump(RF_model, 'RF_model.pkl')
```

```python
param_grid = {'alpha': [0.01, 0.1, 1, 10, 100]}
grid_search = GridSearchCV(L_model, param_grid, cv=5, scoring='r2')
grid_search.fit(X_train_scaled, y_train)
best_L_model = grid_search.best_estimator_
best_L_model_pred= best_L_model.predict(X_test_scaled)
```

```python
best_L_r2 = r2_score(y_test, best_L_model_pred)
best_L_mae = mean_absolute_error(y_test, best_L_model_pred)
best_L_rmse = root_mean_squared_error(y_test, best_L_model_pred)
print(f"R2 Score: {best_L_r2:.4f}")
print(f"MAE: {best_L_mae:.4f}")
print(f"RMSE: {best_L_rmse:.4f}")
```
